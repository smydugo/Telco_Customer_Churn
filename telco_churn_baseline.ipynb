Import libraries for data processing, visualization, and machine learning.


```python
# Step 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
)
```

Load the Telco Customer Churn dataset and preview the first rows.


```python
# Step 2: Load Dataset
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

Inspect dataset shape, column names, data types, and missing values.


```python
# Step 3: Data Understanding
print("Shape:", df.shape)
print("Column names:", list(df.columns))
print("Data types:")
print(df.dtypes)
print("Missing values per column:")
print(df.isnull().sum())
```

Drop customerID (not useful as a feature). Convert TotalCharges to numeric. Remove rows with missing values. Encode Churn as binary (Yes=1, No=0).


```python
# Step 4: Data Cleaning
df = df.drop(columns=["customerID"])
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df = df.dropna()
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
```

Visualize churn distribution to understand class balance.


```python
# Step 5: Exploratory Visualization
sns.countplot(data=df, x="Churn")
plt.tight_layout()
plt.show()
```

Apply one-hot encoding to categorical variables for modeling.


```python
# Step 6: Encoding
df_encoded = pd.get_dummies(df, drop_first=True)
```

Separate input features (X) from the target variable (y).


```python
# Step 7: Feature Split
X = df_encoded.drop(columns=["Churn"])
y = df_encoded["Churn"]
```

Use StandardScaler to normalize features for more stable model training.


```python
# Step 8: Normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Split the data into training (70%) and testing (30%) sets.


```python
# Step 9: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
```

Train logistic regression as a baseline model. Use class_weight='balanced' to reduce bias from class imbalance.


```python
# Step 10: Baseline Model
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, y_train)
```

Generate predictions on the held-out test set.


```python
# Step 11: Predictions
y_pred = model.predict(X_test)
```

Compute accuracy, precision, recall, and F1 on the test set. The classification report provides detailed per-class metrics.


```python
# Step 12: Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

Plot the confusion matrix to analyze prediction quality, including false positives and false negatives.


```python
# Step 13: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
